```python
H = nx.DiGraph()
H.add_nodes_from([2, 3,4,5])
H.add_edge(2, 5)
H.add_edge(2, 4)
H.add_edge(3, 4)
```

```python
P = nx.intersection(G, H)
```

```python
nx.draw_networkx(P, with_labels=True)
```

```python
subH = H.subgraph(P.nodes())
```
